# Polynomial Regression

Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:, 2].values
```

```python
#Fitting the Linear Regression to the dataset (to compair with Polynomial regression)
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
#Fitting Polynomial Linear Regression to the dataset.
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)    # This will 3rd coloum which is sqaure of independent variable.
                                             # and add aditional coloum which contain all the values are 1,
                                             # y = b0 + b1X1 + b2X2, so b0 which is contant treat as 1.
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
X_poly
```

```python
#Visualising the Linear Regression Results.
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('salary')
plt.show()
```

```python
#Visualising the Polynomial Linear Regression results.
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Try Changing the value of degree parameter in PolynomialFeatures "degree = 3"
#Fitting Polynomial Linear Regression to the dataset.
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 3)    # This will 3rd coloum which is sqaure of independent variable. 
                                             # This will 4th coloum which is cube of independent variable.
                                             # and add aditional 1st coloum which contain all the values are 1,
                                             # y = b0 + b1X1 + b2X2 + b3X3, so b0 which is contant treat as 1.
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
X_poly
```

```python
#Visualising the Polynomial Linear Regression results for degree = 3
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Try Changing the value of degree parameter in PolynomialFeatures "degree = 4"
#Fitting Polynomial Linear Regression to the dataset.
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)    # This will 3rd coloum which is sqaure of independent variable. 
                                             # This will 4th coloum which is cube of independent variable.
                                             # This will 5th coloum which is power of 4 of independent variable.
                                             # and add aditional 1st coloum which contain all the values are 1,
                                             # y = b0 + b1X1 + b2X2 + b3X3, so b0 which is contant treat as 1.
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
X_poly
```

```python
#Visualising the Polynomial Linear Regression results for degree = 4
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Creating Smooth curve instead of having line between to points by increasin the resolution.
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
X_grid
```

```python
#Visualising the Polynomial Linear Regression results for degree = 4
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
